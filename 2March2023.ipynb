Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

Scatter plot: A scatter plot is used to visualize the relationship between two variables. It is created using the scatter() function in Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")
plt.title("This is scatter plot")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib is used to create multiple plots in a single figure. It is particularly useful when we want to compare multiple plots or visualize different aspects of the same data in a single figure.

The subplot() function takes three arguments, nrows, ncols, and index, which specify the number of rows, number of columns, and the position of the subplot within the grid, respectively.

Use the following data:

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))


axs[0, 0].plot(x1, y1, color='red')
axs[0, 0].set_title('Line Plot 1')

axs[0, 1].plot(x2, y2, color='blue')
axs[0, 1].set_title('Line Plot 2')

axs[1, 0].plot(x3, y3, color='green')
axs[1, 0].set_title('Line Plot 3')

axs[1, 1].plot(x4, y4, color='orange')
axs[1, 1].set_title('Line Plot 4')
```

```python
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

A bar plot, also known as a bar chart, is a type of graph that represents categorical data with rectangular bars, where the height or length of each bar corresponds to the value or frequency of the category it represents. Bar plots are useful for comparing the relative sizes of different categories or for visualizing how a particular variable changes across categories.

Bar plots can be either vertical or horizontal depending on the orientation of the bars. A vertical bar plot has categories on the x-axis and values on the y-axis, while a horizontal bar plot has categories on the y-axis and values on the x-axis.

To create a bar plot in Matplotlib, we can use the bar() function for vertical bar plots and the barh() function for horizontal bar plots.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```

A box plot is a graphical representation of a dataset that displays the distribution of the data based on five number summary: minimum value, lower quartile (25th percentile), median (50th percentile), upper quartile (75th percentile), and maximum value. The box plot is also known as a box-and-whisker plot, as it consists of a box that represents the interquartile range (IQR) and whiskers that represent the variability outside the IQR.

The box in the plot spans from the lower quartile (Q1) to the upper quartile (Q3), with the median represented as a horizontal line within the box. The length of the box indicates the spread of the middle 50% of the data, and the whiskers extend from the box to the minimum and maximum values, excluding outliers.

Outliers are displayed as individual points outside the whiskers and can indicate extreme values in the data. Some box plots also include notches around the median, which can provide a rough guide to the uncertainty about the median.

Box plots are useful for visualizing the central tendency, variability, and skewness of a dataset, as well as detecting potential outliers and comparing distributions between groups.



```python
plt.boxplot(box1)
```

```python
plt.boxplot(box2)
```
